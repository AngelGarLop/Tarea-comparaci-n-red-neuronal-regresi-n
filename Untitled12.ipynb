# Creación y entrenamiento de los modelos

Preparamos los datos de test y de entrenamiento, para ello primero seperamos la x, en la que metemos todas las columnas menos la columna weekly sales, y la y después los separamos  

```python
# Separar características y variable objetivo
X = df_walmart.drop(['Weekly_Sales'], axis=1)
y = df_walmart['Weekly_Sales']
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)
```

Espcalamos la x y la y

```python
# Escalar las características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Escalar la variable objetivo
scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))
```

Para poder reproducir los datos creamos una semilla

```python
# Fijar la semilla para reproducibilidad
np.random.seed(6)
tf.random.set_seed(6)
```

Definimos el modelo con el que vamos a hacer las pruebas

```python
# Definir el modelo de red neuronal
def build_model():
    model = Sequential()
    model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Hacemos el entrnamiento con los datos sin escalar

```python
# Entrenar y evaluar modelos
model = build_model()
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

Guardamos las métricas para compararlas después

```python
# Evaluar el modelo sin escalar las características
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'MSE sin escalar características: {mse}')
print(f'RMSE sin escalar características: {rmse}')
print(f'MAE sin escalar características: {mae}')
print(f'R2 sin escalar características: {r2}')
```

Hacemos el entrnamiento con los datos de entrada escalados

```python
# Evaluar el modelo con características escaladas
model_scaled = build_model()
model_scaled.fit(X_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

Guardamos las métricas para compararlas despues

```python
y_pred_scaled = model_scaled.predict(X_test_scaled)
mse_scaled = mean_squared_error(y_test, y_pred_scaled)
rmse_scaled = np.sqrt(mse_scaled)
mae_scaled = mean_absolute_error(y_test, y_pred_scaled)
r2_scaled = r2_score(y_test, y_pred_scaled)
print(f'MSE con características escaladas: {mse_scaled}')
print(f'RMSE con características escaladas: {rmse_scaled}')
print(f'MAE con características escaladas: {mae_scaled}')
print(f'R2 con características escaladas: {r2_scaled}')
```

Hacemos el entrnamiento con los datos de entrada y salida escalados

```python
# Evaluar el modelo con salida escalada
model_output_scaled = build_model()
model_output_scaled.fit(X_train_scaled, y_train_scaled, epochs=100, batch_size=32, validation_split=0.2)
```

Desescalamos la salida y guardamos las métricas para compararlas después

```python

y_pred_output_scaled = model_output_scaled.predict(X_test_scaled)
y_pred_output = scaler_y.inverse_transform(y_pred_output_scaled)
mse_output_scaled = mean_squared_error(y_test, y_pred_output)
rmse_output_scaled = np.sqrt(mse_output_scaled)
mae_output_scaled = mean_absolute_error(y_test, y_pred_output)
r2_output_scaled = r2_score(y_test, y_pred_output)
print(f'MSE con salida escalada: {mse_output_scaled}')
print(f'RMSE con salida escalada: {rmse_output_scaled}')
print(f'MAE con salida escalada: {mae_output_scaled}')
print(f'R2 con salida escalada: {r2_output_scaled}')
```

Preparamos el modelo  que se va a usar con los hyperparametros en el que vamos a probar entre 20 y 10 capas, entre 16 y 64 neuronas por capa, relu, tanh y sigmoid de activación y entre 0.02 y 0.002 de leraning rate 

```python
def build_model_optimizado(hp):
    model = Sequential()

    # Definir número de capas
    for i in range(hp.Int('num_layers', 2, 10)):  # Probar entre 2 y 5 capas
        model.add(Dense(units=hp.Int(f'units_{i}', min_value=16, max_value=64, step=16),
                        activation=hp.Choice('activation', values=['relu', 'tanh', 'sigmoid'])))

    model.add(Dense(1, activation='sigmoid'))  # Capa de salida

    # Definir el optimizador con el learning rate como hiperparámetro
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='LOG')),
                  loss='mean_squared_error',
                  metrics=['accuracy','mean_absolute_error'])
    return model
```

Cargamos el GridSearch con el objetivo del menor error absoluto medio y 100 pruebas

```python
# Optimización de hiperparámetros
tuner = kt.GridSearch(
    build_model_optimizado,  # La función que crea el modelo
    objective='val_mean_absolute_error',  # La métrica que se optimiza
    max_trials=200,  # Número máximo de combinaciones de hiperparámetros
    executions_per_trial=1,  # Número de veces que se ejecutará cada prueba
    directory='my_dir',  # Directorio para guardar los resultados
    project_name='classWallmart_grid'  # Nombre del proyecto
)
```

Ejecutamos la busqueda del mejor modelo con los hiperparametros

```python
# Ejecutar la búsqueda
tuner.search(X_train_scaled, y_train_scaled, epochs=100, validation_data=(X_test_scaled, y_test_scaled))
```

Creamos el modelo con los mejores hiperparametros 

```python
# Obtener los mejores hiperparámetros
best_hps = tuner.get_best_hyperparameters()[0]
print(f"Mejores hiperparámetros: {best_hps.values}")

# Crear el modelo final con los mejores hiperparámetros
model_grid = tuner.hypermodel.build(best_hps)
```

Entrenamos el modelo

```python
# Entrenar el modelo final
model_grid.fit(X_train_scaled, y_train_scaled, epochs=100, validation_data=(X_test_scaled, y_test_scaled))
```

Desescalamos la salida y guardamos las métricas para compararlas después

```python
# Evaluar el modelo optimizado
best_model = model_grid
y_pred_optimized_scaled = best_model.predict(X_test_scaled)
y_pred_optimized = scaler_y.inverse_transform(y_pred_optimized_scaled.reshape(-1, 1))
mse_optimized = mean_squared_error(y_test, y_pred_optimized)
rmse_optimized = np.sqrt(mse_optimized)
mae_optimized = mean_absolute_error(y_test, y_pred_optimized)
r2_optimized = r2_score(y_test, y_pred_optimized)
print(f'MSE del modelo optimizado con grid: {mse_optimized}')
print(f'RMSE del modelo optimizado con grid: {rmse_optimized}')
print(f'MAE del modelo optimizado con grid: {mae_optimized}')
print(f'R2 del modelo optimizado con grid: {r2_optimized}')
```

Cargamos el BayesianOptimization con el objetivo del menor error absoluto medio y 100 pruebas

```python
tuner_baye = kt.BayesianOptimization(
    build_model_optimizado,  # La función que crea el modelo
    objective='val_mean_absolute_error',  # La métrica que se optimiza
    max_trials=200,  # Número máximo de combinaciones de hiperparámetros
    executions_per_trial=1,  # Número de veces que se ejecutará cada prueba
    directory='my_dir',  # Directorio para guardar los resultados
    project_name='classWallmart_baye'  # Nombre del proyecto
)
```

Ejecutamos la busqueda del mejor modelo con los hiperparametros

```python
tuner_baye.search(X_train_scaled, y_train_scaled, epochs=100, validation_data=(X_test_scaled, y_test_scaled))
```

```python
# Mostrar los mejores hiperparámetros encontrados
best_hps = tuner_baye.get_best_hyperparameters(num_trials=1)[0]
print(f"Mejores hiperparámetros: {best_hps.values}")
model_baye = tuner.hypermodel.build(best_hps)
```

Entrenamos el modelo

```python
model_baye.fit(X_train_scaled, y_train_scaled, epochs=100, validation_data=(X_test_scaled, y_test_scaled))
```

```python
# Evaluar el modelo optimizado
best_model = model_baye
y_pred_optimized_scaled_baye = best_model.predict(X_test_scaled)
y_pred_optimized_baye = scaler_y.inverse_transform(y_pred_optimized_scaled_baye.reshape(-1, 1))
mse_optimized_baye = mean_squared_error(y_test, y_pred_optimized_baye)
rmse_optimized_baye = np.sqrt(mse_optimized_baye)
mae_optimized_baye = mean_absolute_error(y_test, y_pred_optimized_baye)
r2_optimized_baye = r2_score(y_test, y_pred_optimized_baye)
print(f'MSE del modelo optimizado bayesiano: {mse_optimized_baye}')
print(f'RMSE del modelo optimizado bayesiano: {rmse_optimized_baye}')
print(f'MAE del modelo optimizado bayesiano: {mae_optimized_baye}')
print(f'R2 del modelo optimizado bayesiano: {r2_optimized_baye}')
```

```python
# Gráficos de dispersión para cada modelo
def plot_predictions(y_test, y_pred, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test, y_pred, color="red", alpha=0.5, label="Predicciones")
    plt.scatter(y_test, y_test, color="blue", alpha=0.5, label="Valores Reales")
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="black", linestyle="--", label="Línea Perfecta")
    plt.xlabel("Valores reales")
    plt.ylabel("Predicciones")
    plt.title(title)
    plt.legend()
    plt.show()
```
